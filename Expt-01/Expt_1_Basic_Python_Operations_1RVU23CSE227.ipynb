```python

import numpy as np

a = np.array([[1, 2], [4, 5]])
b = np.array([[1, 2], [4, 5]])

while True:
    print("1. Add\n2. Subtract\n3. Multiply\n4. Divide\n5. Dot product\n6. Exponentiation\n7. Logarithm\n8. Natural logarithm\n9. Exit")
    n = int(input("Enter the option number: "))

    if n < 1 or n > 9:
        print("No such option exists, please enter a valid option.\n")
        continue

    if n == 1:
        c = np.add(a, b)
        print("Sum\n", c)
        print("\n")

    elif n == 2:
        d = np.subtract(a, b)
        print("Difference\n", d)
        print("\n")

    elif n == 3:
        e = np.multiply(a, b)
        print("Product\n", e)
        print("\n")

    elif n == 4:
        f = np.divide(a, b)
        print("Division\n", f)
        print("\n")

    elif n == 5:
        g = np.dot(a, b)
        print("Dot product\n", g)
        print("\n")

    elif n == 6:
        h, i = np.exp(a), np.exp(b)
        print("Exponentiation for array a : \n", h)
        print("Exponentiation for array b : \n", i)
        print("\n")

    elif n==7:
        l, m = np.log10(a), np.log10(b)
        print("Logarithm for array a : \n",l)
        print("Logarithm for array b : \n",m)

    elif n == 8:
        x, y =np.log10(a), np.log10(b)
        print("Natural Logarithm for array a : \n",x)
        print("Natural Logarithm for array b : \n",y)
    elif n == 9:
        break

    else:
        print("No such option exist, Please enter existing options.\n")
```

```python
import numpy as np
import matplotlib.pyplot as mpl

x = np.random.randn(50)
y1 = x * 5 + 3
y2 = -5 * x
y3 = np.random.randn(50)

mpl.scatter(x, y1, color="green", label="Positive correlation")
mpl.scatter(x, y2, color="red", label="Negative correlation")
mpl.scatter(x, y3, color="blue", label="Zero correlation")

mpl.rcParams.update({'figure.figsize': (10, 8), 'figure.dpi': 100})

mpl.xlabel("X-axis")
mpl.ylabel("Y-axis")
mpl.title("Correlation and Scatter plots")
mpl.legend()
mpl.show()
```

```python
import numpy as np
import matplotlib.pyplot as mpl

x = np.random.randn(50)
y1 = x * 5 + 3
y2 = -5 * x
y3 = np.random.randn(50)

corrcoef = np.round(np.corrcoef(x, y1)[0, 1], 1)

mpl.scatter(x, y1, color="green", label=f"Positive corr coefficient = {corrcoef}")
mpl.scatter(x, y2, color="red", label=f"Negative corr coefficient = {corrcoef}")
mpl.scatter(x, y3, color="blue", label=f"Zero corr coefficient = {corrcoef}")

mpl.rcParams.update({'figure.figsize': (10, 8), 'figure.dpi': 100})

mpl.xlabel("X-axis")
mpl.ylabel("Y-axis")
mpl.title("Correlation and Scatter plots")
mpl.legend()
mpl.show()
```
